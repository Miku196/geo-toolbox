# geo-toolbox Jupyter Demo

Browser-grade GIS operations in a notebook. Powered by Rust via PyO3.

Install: `pip install geo-toolbox geo-toolbox-magic`

```python
%load_ext geo_toolbox_magic
```

## List all available tools

```python
%geo_list
```

## Geohash — Encode & Decode

```python
%%geo geohash_encode
{"lat": 39.9042, "lon": 116.4074, "precision": 8}
```

```python
%geo geohash_decode {"hash": "wx4g0ec1"}
```

## Tile Math — Lat/Lon to Tile Coordinates

```python
%%geo latlon_to_tile
{"lat": 39.9042, "lon": 116.4074, "zoom": 10}
```

## Area Calculation (Beijing Forbidden City bounding box)

```python
%%geo compute_area_sqm
{"geojson_geom": "{\"type\":\"Polygon\",\"coordinates\":[[[116.38,39.90],[116.40,39.90],[116.40,39.92],[116.38,39.92],[116.38,39.90]]]}"}
```

## CRS Transform — WGS84 to GCJ-02

```python
%%geo crs_transform
{"coords": [{"x": 116.4074, "y": 39.9042}], "from_crs": "EPSG:4326", "to_crs": "EPSG:4490"}
```

## Tool Schema Inspection
